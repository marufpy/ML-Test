<h1>Code #1: Filling null values with a single value</h1>

```python
dict = {'First Score':[100,90,np.nan,95],
       'Second Score': [30,45,56,np.nan],
       'Third Score':[np.nan,40,80,98]}
```

```python
df = pd.DataFrame(dict)
```

```python
df.fillna(0)
```

<h1>Code #2: Filling null values with the previous ones
</h2>

```python
dict = {'First Score':[100,90,np.nan,95],
       'Second Score': [30,45,56,np.nan],
       'Third Score':[np.nan,40,80,98]}
```

```python
df = pd.DataFrame(dict)
```

```python
df.fillna(method='pad')
```

<h1>Code #3: Filling null value with the next ones</h1>

```python
dict = {'First Score':[100,90,np.nan,95],
       'Second Score': [30,45,56,np.nan],
       'Third Score':[np.nan,40,80,98]}
df = pd.DataFrame(dict)
```

```python
df.fillna(method='bfill')
```

<h1>Code #6: Using interpolate() function to fill the missing values using linear method.
</h1>

```python
df = pd.DataFrame({'A':[12,4,5,None,1],
                  'B':[None,2,54,3,None],
                  'C':[20,16,None,3,8],
                  'D':[14,3,None,None,6]})
```

```python
df
```

```python
df.interpolate(method='linear',limit_direction='forward')
```

```python
df.interpolate(method='polynomial',order =2)
```

```python
df
```
